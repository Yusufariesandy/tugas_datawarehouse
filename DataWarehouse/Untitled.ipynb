```python
c.executescript('drop table if exists kota')
c.executescript('drop table if exists pembeli')
c.executescript('drop table if exists produk')
c.executescript('drop table if exists payment')
c.executescript('drop table if exists transaksi')
c.execute("CREATE TABLE IF NOT EXISTS 'kota' ('id_kota' INT, 'nama_kota' VARCHAR(255) NULL, PRIMARY KEY ('id_kota'))")
c.execute("CREATE TABLE IF NOT EXISTS 'payment' ('id_payment' INT, 'jenis' VARCHAR(255) NULL, PRIMARY KEY ('id_payment'))")
c.execute("CREATE TABLE IF NOT EXISTS 'produk' ('id_produk' INT, 'kategori' VARCHAR(255) NULL, 'harga' FLOAT NULL, PRIMARY KEY ('id_produk'))")
c.execute("CREATE TABLE IF NOT EXISTS 'pembeli' ('id_pembeli' INT, 'invoice_id' VARCHAR(255) NULL, 'type_pembeli' VARCHAR(255) NULL, 'gender' VARCHAR(255) NULL, 'kota_id' INT NOT NULL, PRIMARY KEY ('id_pembeli'), FOREIGN KEY ('kota_id') REFERENCES 'kota' ('id_kota') ON DELETE NO ACTION ON UPDATE NO ACTION)")
c.execute("CREATE TABLE IF NOT EXISTS 'transaksi' ('id_transaksi' INT, 'tanggal' DATE NULL, 'jumlah' INT NULL, 'total' FLOAT NULL, 'pembeli_id' INT NOT NULL, 'produk_id' INT NOT NULL, 'payment_id' INT NOT NULL, PRIMARY KEY ('id_transaksi'), FOREIGN KEY ('pembeli_id') REFERENCES 'pembeli' ('id_pembeli') ON DELETE NO ACTION ON UPDATE NO ACTION, FOREIGN KEY ('produk_id') REFERENCES 'produk' ('id_produk') ON DELETE NO ACTION ON UPDATE NO ACTION, FOREIGN KEY ('payment_id') REFERENCES 'payment' ('id_payment') ON DELETE NO ACTION ON UPDATE NO ACTION)")
```

```python
data = pd.read_excel("supermarket_sales.xlsx")
data
```

```python
kota = []
produk = []
payment = []
pembeli = []
for i,j in data.iterrows():
    if j[1] not in kota:
        kota.append(j[1])
        
    k = -1
    for index, value in enumerate(kota):
        if value == j[1]:
            k = index+1
            break
        
    if[j[0], j[2], j[3], k] not in pembeli:
        pembeli.append([j[0], j[2], j[3], k])
```

```python
c.execute("DELETE FROM KOTA")
for index, i in enumerate(kota):
    c.execute("INSERT INTO KOTA VALUES ({},'{}')".format(index, i))

c.execute("DELETE FROM PRODUK")
for index, i in enumerate(produk):
    c.execute("INSERT INTO PRODUK VALUES ({},'{}','{}')".format(index, i[0],i[1]))

c.execute("DELETE FROM PAYMENT")
for index, i in enumerate(payment):
    c.execute("INSERT INTO PAYMENT VALUES ({},'{}')".format(index, i))

c.execute("DELETE FROM PEMBELI")
for index, i in enumerate(pembeli):
    c.execute("INSERT INTO PEMBELI VALUES ({},'{}','{}','{}','{}')".format(index, i[0],i[1],i[2],i[3]))    

c.execute("DELETE FROM TRANSAKSI")
for index, i in enumerate(transaksi):
    c.execute("INSERT INTO TRANSAKSI VALUES ({},'{}','{}','{}','{}','{}','{}')".format(index, i[0],i[1],i[2],i[3],i[4],i[5]))
```
